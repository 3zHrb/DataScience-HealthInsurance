# Data Processing for Machine Learning Model

## - Separating Data into X and Y

```python
X = df[['age', 'bmi', 'sex', 'smoker', 'region', 'children']]
Y = df['charges']
```

## - Scaling **age** and **bmi**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X[['age', 'bmi']])
transformedAgeAndBMI = scaler.transform(X[['age', 'bmi']])
transformedAgeAndBMI
```

```python
X.loc[:, ['age', 'bmi']] = transformedAgeAndBMI
```

```python
scaler.transform(np.array([[19.0, 27.9]]))
```

```python
X
```

## - Encoding **sex**, **smoker** and **region**

```python
EncodingSmokerAndMerging = pd.merge(X,pd.get_dummies(X['smoker']), left_index=True, right_index=True)
EncodingSmokerAndMerging
```

```python
EncodingSexAndMerging = pd.merge(EncodingSmokerAndMerging,pd.get_dummies(X['sex']), left_index=True, right_index=True)
EncodingSexAndMerging
```

```python
AllEncodedObjectTypeMerged = pd.merge(EncodingSexAndMerging,pd.get_dummies(X['region']), left_index=True, right_index=True)
AllEncodedObjectTypeMerged
```

```python
## to mitigate correlation we will do the following
## ['smoker'] == Yes column is dropped
## ['sex'] == female will be dropped
## ['region'] == southwest will be dropped
## X_Data_ML will be a data that is ready for the ML model (not for analysing)
X_Data_ML = AllEncodedObjectTypeMerged
X_Data_ML = X_Data_ML.drop(columns = ['sex', 'smoker', 'region', 'yes', 'female', 'southwest'])
X_Data_ML
```

## - Splitting X and Y into training and testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_Data_ML, Y)
```

## - Importing all regression models

```python
from sklearn.linear_model import LinearRegression, BayesianRidge, SGDRegressor
from sklearn.svm import LinearSVR
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
import xgboost as xgb
```

## - Importing Cross Validation

```python
from sklearn.model_selection import cross_val_score
```

```python
cv_models = {
'LinearRegression': LinearRegression,
'BayesianRidge': BayesianRidge,
'SGDRegressor': SGDRegressor,
'LinearSVR': LinearSVR,
'DecisionTreeRegressor': DecisionTreeRegressor,
'xgb.XGBRegressor': xgb.XGBRegressor,
'RandomForestRegressor': RandomForestRegressor,
'KNeighborsRegressor': KNeighborsRegressor,
'GradientBoostingRegressor': GradientBoostingRegressor,
'LinearSVR': LinearSVR,
'SVR': SVR
}

for modelName, model in cv_models.items():
    print(f'model name is: ({modelName}) average cv model score is: (%{np.round(cross_val_score(model(), X_Data_ML, Y, cv=5).mean() * 100, decimals=2)})')
```

#### Result:
###### The best 3 models with default parameters are:
###### xgb.XGBRegressor (%80.75)
###### RandomForestRegressor (%83.49)
###### GradientBoostingRegressor (%85.50)

###### the result is still not good, so we will try to improve it using two techniques:
###### 1- PCA 
###### 2- Parameter Tuning using GridSearchCV

###### The 3 models that will be selected for these two further enhancment techniques are the best 3 models (**xgb.XGBRegressor**, **RandomForestRegressor**, **GradientBoostingRegressor**)

## - Principle Components Analysis

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X_Data_ML)
NewPCA = pca.transform(X_Data_ML)
NewPCA_DF = pd.DataFrame(NewPCA)
```

```python
NewPCA_DF
```

```python
for modelName, model in cv_models.items():
    print(f'model name is: ({modelName}) average cv model score is: (%{np.round(cross_val_score(model(), NewPCA_DF, Y, cv=5).mean() * 100, decimals=2)})')
```

#### Result: 
###### we can see that using PCA with two PCA features did not improve the result but it effected the result negatively 

## - Using RandomizedSearchCV For Parameter Tuning

```python
from sklearn.model_selection import RandomizedSearchCV #, GridSearchCV
```

```python
## models (**xgb.XGBRegressor**, **RandomForestRegressor**, **GradientBoostingRegressor**)

## total combinations for all models added up 1,512
modelsAndParametersForRandomizedSearchCV = {
    
    #    ## 960 total combinations
    GradientBoostingRegressor(): {
        'loss': ['squared_error', 'absolute_error', 'huber', 'quantile'],
        'learning_rate': [0.01,0.2],
        'min_samples_split': [3,5,7,10],
        'n_estimators': [250, 500],
        'max_depth': [5,10,15,20,25,30]
    },
    
    ## 360 total combinations
    RandomForestRegressor(): {
        'n_estimators': [250, 500],
        'max_depth': [5,10,15,20,25,30],
        'min_samples_leaf': [2,3,5],
        'min_samples_split': [3,5,7,10],
        'criterion': ['squared_error']
    },
    
    ## 960 total combinations
    xgb.XGBRegressor(): {
        'max_depth': [3,5,7,10],
        'learning_rate': [0.01, 0.1, 0.2],
        'n_estimators': [200, 300, 350, 500],
        'gamma': [0, 0.1],
        'min_child_weight': [1,5,7,10],
        'subsample': [0.5, 1]
    }
    
}

bestModels = []

for model, modelParameter in modelsAndParametersForRandomizedSearchCV.items():
    print(f'model: {model}, Parameter: {modelParameter}')
    RandomizedSCV = RandomizedSearchCV(model, modelParameter, cv=5, verbose=10)
    RandomizedSCV.fit(X_Data_ML, Y)
    bestModels.append({
        'Model': model,
        'Best Parameters': RandomizedSCV.best_params_,
        'Best Score': RandomizedSCV.best_score_
    })
    
    
```

```python
bestModels
```

#### Result:
###### result **before** using RandomizedSearchCV
###### xgb.XGBRegressor (%80.75) 
###### RandomForestRegressor (%83.49)
###### GradientBoostingRegressor (%85.50)

###### result **after** using RandomizedSearchCV

###### xgb.XGBRegressor (%86.17) did better by almost 6% using RandomizedSearchCV
###### RandomForestRegressor (%86.07) did better by almost %2.5 using RandomizedSearchCV
###### GradientBoostingRegressor (%85.08) did slightly wores by %0.42 using RandomizedSearchCV

###### so the best model so far will be xgb.XGBRegressor with the following parameters:
###### {'subsample': 1,
######    'n_estimators': 500,
######    'min_child_weight': 10,
######    'max_depth': 3,
######    'learning_rate': 0.01,
######    'gamma': 0},
######   'Best Score': 0.8617365879670681}]

## - Using Cross Validation again after RandomizedSearchCV to make sure that the best model we selected has low variance
